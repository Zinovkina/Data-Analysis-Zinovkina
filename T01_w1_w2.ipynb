````python
import math

circumference = float(input("–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω—É —ç–∫–≤–∞—Ç–æ—Ä–∏–∞–ª—å–Ω–æ–π –ø–∞—Ä–∞–ª–ª–µ–ª–∏ —à–∞—Ä–∞: "))
radius = circumference / (2 * math.pi)
volume = (4/3) * math.pi * radius ** 3
print(f"–ü—Ä–∏ –¥–ª–∏–Ω–µ —ç–∫–≤–∞—Ç–æ—Ä–∏–∞–ª—å–Ω–æ–π –ø–∞—Ä–∞–ª–ª–µ–ª–∏ {circumference:.2f} –æ–±—ä–µ–º —à–∞—Ä–∞ —Ä–∞–≤–µ–Ω {volume:.2f}")
````

````python
#–ò–º–µ—é—Ç—Å—è –¥–≤–∞  ùëõ‚Äì–º–µ—Ä–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞  ùë• –∏  ùë¶, –∫–æ—Ç–æ—Ä—ã–µ –∑–∞–¥–∞—é—Ç –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã  ùëõ —Ç–æ—á–µ–∫ –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ (—Å–ª—É—á–∞–π–Ω—ã–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞). –ù–∞–π—Ç–∏ –Ω–∞–∏–±–æ–ª–µ–µ –±–ª–∏–∑–∫–∏–µ –¥—Ä—É–≥ –¥—Ä—É–≥—É —Ç–æ—á–∫–∏.

import random

def distance (a,b):
    return ((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2) ** 0.5
n = 5
x, y = [], []

for i in range(n):
    x.append(random.randint(1,100))
    y.append(random.randint(1,100))
    
closest_pair = None
min_distance = float('inf')

for i in range(n):
    for j in range(i+1, n):
        d = distance ([x[i], y[i]], [x[j], y[j]])
        if d < min_distance:
            min_distance, closest_pair = d, (i, j)
            
print("–ë–ª–∏–∂–∞–π—à–∏–µ —Ç–æ—á–∫–∏:", closest_pair)
print("–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –Ω–∏–º–∏:", min_distance)
````

````python
#–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –∑–∞–¥–∞–Ω –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ —Å–≤–æ–∏—Ö –≤–µ—Ä—à–∏–Ω:  (ùë•1,ùë¶1),(ùë•2,ùë¶2),(ùë•3,ùë¶3). –ó–Ω–∞—á–µ–Ω–∏—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å –ø–æ–º–æ—â—å—é –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏—è. –û–Ω–∏ –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–µ—Ü–µ–ª—ã–º–∏. –ù–∞–π—Ç–∏ –ø–µ—Ä–∏–º–µ—Ç—Ä –∏ –ø–ª–æ—â–∞–¥—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞.

import math

x1, y1 = 1, 0
x2, y2 = 3, 0
x3, y3 = 0, 4
a = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
b = math.sqrt((x3 - x2) ** 2 + (y3 - y2) ** 2)
c = math.sqrt((x1 - x3) ** 2 + (y1 - y3) ** 2)
p = (a + b + c) / 2
s = math.sqrt(p * (p - a) * (p - b) * (p - c))
print("–î–ª–∏–Ω—ã —Å—Ç–æ—Ä–æ–Ω —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞: a =", a, "b =", b, "c =", c)
print("–ü–µ—Ä–∏–º–µ—Ç—Ä —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞:", a + b + c)
print("–ü–ª–æ—â–∞–¥—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞:", s)
````

````python
#–î–ª—è –≤–≤–µ–¥–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ –≤—ã–≤–µ–¥–∏—Ç–µ (–Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ):
#–≤—Ç–æ—Ä–æ–π —Å–∏–º–≤–æ–ª —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–∏;
#–ø—Ä–µ–¥–ø–æ—Å–ª–µ–¥–Ω–∏–π —Å–∏–º–≤–æ–ª —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–∏;
#–≤—Å—é —Å—Ç—Ä–æ–∫—É, –∫—Ä–æ–º–µ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –¥–≤—É—Ö —Å–∏–º–≤–æ–ª–æ–≤;
#–≤—Å–µ —Å–∏–º–≤–æ–ª—ã —Å —á–µ—Ç–Ω—ã–º–∏ –∏–Ω–¥–µ–∫—Å–∞–º–∏ (—Å—á–∏—Ç–∞—è, —á—Ç–æ –∏–Ω–¥–µ–∫—Å–∞—Ü–∏—è –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å 0, –ø–æ—ç—Ç–æ–º—É —Å–∏–º–≤–æ–ª—ã –≤—ã–≤–æ–¥—è—Ç—Å—è, –Ω–∞—á–∏–Ω–∞—è —Å –ø–µ—Ä–≤–æ–≥–æ);
#–≤—Å–µ —Å–∏–º–≤–æ–ª—ã —Å –Ω–µ—á–µ—Ç–Ω—ã–º–∏ –∏–Ω–¥–µ–∫—Å–∞–º–∏, —Ç–æ –µ—Å—Ç—å –Ω–∞—á–∏–Ω–∞—è —Å–æ –≤—Ç–æ—Ä–æ–≥–æ —Å–∏–º–≤–æ–ª–∞ —Å—Ç—Ä–æ–∫–∏;
#–≤—Å–µ —Å–∏–º–≤–æ–ª—ã –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ;
#–≤—Å–µ —Å–∏–º–≤–æ–ª—ã —Å—Ç—Ä–æ–∫–∏ —á–µ—Ä–µ–∑ –æ–¥–∏–Ω –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ, –Ω–∞—á–∏–Ω–∞—è —Å –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ;
#–¥–ª–∏–Ω—É –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏.

s = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É: ")
second = s[1]
print("–í—Ç–æ—Ä–æ–π —Å–∏–º–≤–æ–ª:", second)
penultimate = s[-2]
print("–ü—Ä–µ–¥–ø–æ—Å–ª–µ–¥–Ω–∏–π —Å–∏–º–≤–æ–ª:", penultimate)
without_last_two = s[:-2]
print("–°—Ç—Ä–æ–∫–∞ –±–µ–∑ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –¥–≤—É—Ö —Å–∏–º–≤–æ–ª–æ–≤:", without_last_two)
even = s[::2]
print("–°–∏–º–≤–æ–ª—ã —Å —á–µ—Ç–Ω—ã–º–∏ –∏–Ω–¥–µ–∫—Å–∞–º–∏:", even)
odd = s[1::2]
print("–°–∏–º–≤–æ–ª—ã —Å –Ω–µ—á–µ—Ç–Ω—ã–º–∏ –∏–Ω–¥–µ–∫—Å–∞–º–∏:", odd)
reverse = s[::-1]
print("–°–∏–º–≤–æ–ª—ã –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ:", reverse)
reverse_through = s[::-2]
print("–°–∏–º–≤–æ–ª—ã —á–µ—Ä–µ–∑ –æ–¥–∏–Ω –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ:", reverse_through)
length = len(s)
print("–î–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏:", length)
````

````python
#–ù–µ –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ count, –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ –≤—ã–ø–æ–ª–Ω–∏—Ç–µ:
#–µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤ –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç, –≤—ã–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç ¬´–Ω–µ—Ç —Å–∏–º–≤–æ–ª–∞¬ª;
#–µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–µ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –∏–Ω–¥–µ–∫—Å;
#–µ—Å–ª–∏ —Å–∏–º–≤–æ–ª * –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –¥–≤–∞ –∏ –±–æ–ª–µ–µ —Ä–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ –∏–Ω–¥–µ–∫—Å –µ–≥–æ –ø–µ—Ä–≤–æ–≥–æ, –≤—Ç–æ—Ä–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è, —É–¥–∞–ª–∏—Ç–µ –ø–µ—Ä–≤—ã–π –∏ –ø–æ—Å–ª–µ–¥–Ω–∏–π —Å–∏–º–≤–æ–ª * –∏–∑ —Å—Ç—Ä–æ–∫–∏.

s = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É: ")
star_indexes = []
for i in range(len(s)):
    if s[i] == '*':
        star_indexes.append(i)

if len(star_indexes) == 0:
    print("–Ω–µ—Ç —Å–∏–º–≤–æ–ª–∞")
elif len(star_indexes) == 1:
    print("–ò–Ω–¥–µ–∫—Å —Å–∏–º–≤–æ–ª–∞ *:", star_indexes[0])
else:
    print("–ò–Ω–¥–µ–∫—Å –ø–µ—Ä–≤–æ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è *:", star_indexes[0])
    print("–ò–Ω–¥–µ–∫—Å –≤—Ç–æ—Ä–æ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è *:", star_indexes[1])
    print("–ò–Ω–¥–µ–∫—Å –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è *:", star_indexes[-1])
    s = s.replace('*', '', 1)  
    s = s[::-1].replace('*', '', 1)[::-1] 
    print("–°—Ç—Ä–æ–∫–∞ –±–µ–∑ –ø–µ—Ä–≤–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —Å–∏–º–≤–æ–ª–∞ *:", s)
````

````python
#–ü–æ –Ω–æ–º–µ—Ä—É –≥–æ–¥–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω—ã–π –≥–æ–¥ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º. (–≥–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º, –µ—Å–ª–∏ –µ–≥–æ –Ω–æ–º–µ—Ä –∫—Ä–∞—Ç–µ–Ω 4, –Ω–æ –Ω–µ –∫—Ä–∞—Ç–µ–Ω 100, –∞ —Ç–∞–∫–∂–µ –µ—Å–ª–∏ –æ–Ω –∫—Ä–∞—Ç–µ–Ω 400).

year = int(input("–í–≤–µ–¥–∏—Ç–µ –≥–æ–¥: "))
if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    result = True
else:
    result = False

if result:
    print(year, "–≥–æ–¥ - –≤–∏—Å–æ–∫–æ—Å–Ω—ã–π!")
else:
    print(year, "–≥–æ–¥ - –Ω–µ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–π.")
````

````python
#–î–∞–Ω—ã —Ç–æ—á–∫–∏  ùê¥=(1,1),ùêµ=(5,5),ùê∂=(1,4),ùê∑=(7,2). –ù–∞–π—Ç–∏ —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –æ—Ç—Ä–µ–∑–∫–æ–≤  ùê¥ùêµ –∏  ùê∂ùê∑.

A = [1, 1]
B = [5, 5]
C = [1, 4]
D = [7, 2]
x1, y1 = A
x2, y2 = B
x3, y3 = C
x4, y4 = D
denominator = (x1 - x2) * (y3 - y4) - (y1 - y2) * (x3 - x4)
dx_ab = x1 * y2 - y1 * x2
dx_cd = x3 * y4 - y3 * x4
if denominator == 0:
    result = "–û—Ç—Ä–µ–∑–∫–∏ AB –∏ CD –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã –∏ –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è."
else:
    x = (dx_ab * (x3 - x4) - (x1 - x2) * dx_cd) / denominator
    y = (dx_ab * (y3 - y4) - (y1 - y2) * dx_cd) / denominator
    result = [x, y]
    
print("–¢–æ—á–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è AB –∏ CD:", result)
````
