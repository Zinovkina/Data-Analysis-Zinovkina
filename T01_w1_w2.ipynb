````python
#Имеются два  𝑛–мерных вектора  𝑥 и  𝑦, которые задают координаты  𝑛 точек на плоскости (случайные целые числа). Найти наиболее близкие друг другу точки.

import random

def distance (a,b):
    return ((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2) ** 0.5
n = 5
x, y = [], []

for i in range(n):
    x.append(random.randint(1,100))
    y.append(random.randint(1,100))
    
closest_pair = None
min_distance = float('inf')

for i in range(n):
    for j in range(i+1, n):
        d = distance ([x[i], y[i]], [x[j], y[j]])
        if d < min_distance:
            min_distance, closest_pair = d, (i, j)
            
print("Ближайшие точки:", closest_pair)
print("Расстояние между ними:", min_distance)
````

````python
#Не используя метод count, для заданной строки выполните:
#если символ * в данной строке отсутствует, выведите текст «нет символа»;
#если символ * встречается в строке только один раз, выведите его индекс;
#если символ * встречается два и более раз, выведите индекс его первого, второго и последнего вхождения, удалите первый и последний символ * из строки.

s = input("Введите строку: ")
star_indexes = []
for i in range(len(s)):
    if s[i] == '*':
        star_indexes.append(i)

if len(star_indexes) == 0:
    print("нет символа")
elif len(star_indexes) == 1:
    print("Индекс символа *:", star_indexes[0])
else:
    print("Индекс первого вхождения *:", star_indexes[0])
    print("Индекс второго вхождения *:", star_indexes[1])
    print("Индекс последнего вхождения *:", star_indexes[-1])
    s = s.replace('*', '', 1)  
    s = s[::-1].replace('*', '', 1)[::-1] 
    print("Строка без первого и последнего символа *:", s)
````

````python
#По номеру года определите, является ли данный год високосным. (год является високосным, если его номер кратен 4, но не кратен 100, а также если он кратен 400).

year = int(input("Введите год: "))
if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    result = True
else:
    result = False

if result:
    print(year, "год - високосный!")
else:
    print(year, "год - не високосный.")
````

````python
#Даны точки  𝐴=(1,1),𝐵=(5,5),𝐶=(1,4),𝐷=(7,2). Найти точку пересечения отрезков  𝐴𝐵 и  𝐶𝐷.

A = [1, 1]
B = [5, 5]
C = [1, 4]
D = [7, 2]
x1, y1 = A
x2, y2 = B
x3, y3 = C
x4, y4 = D
denominator = (x1 - x2) * (y3 - y4) - (y1 - y2) * (x3 - x4)
dx_ab = x1 * y2 - y1 * x2
dx_cd = x3 * y4 - y3 * x4
if denominator == 0:
    result = "Отрезки AB и CD параллельны и не пересекаются."
else:
    x = (dx_ab * (x3 - x4) - (x1 - x2) * dx_cd) / denominator
    y = (dx_ab * (y3 - y4) - (y1 - y2) * dx_cd) / denominator
    result = [x, y]
    
print("Точка пересечения AB и CD:", result)
````
